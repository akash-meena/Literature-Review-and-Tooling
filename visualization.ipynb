```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
df = pd.read_csv('bc5a3e94-f69c-4d6a-887d-5b8044dd11f6.csv')
```

```python
df.head()
```

```python
data = df.iloc[36].tolist()
```

```python
data = data[1:]
```

```python
year = np.arange(2004,2016,1)
```

```python
plt.rcParams["figure.figsize"] = [12, 8]
    
X_Y_Spline = make_interp_spline(year, data)
X_ = np.linspace(year.min(), year.max(), 500)
Y_ = X_Y_Spline(X_)
plt.plot(X_, Y_)

plt.xlabel('Year')
plt.scatter(year,data)
plt.ylabel('Length of National Highways in KMs')
plt.title('Year wise All India Length of National Highways from 2004 to 2015')
plt.savefig('Highway',dpi=1000)
```

```python
df1 = pd.read_csv('india-gdp-gross-domestic-product.csv')
df2 = pd.read_csv('india-unemployment-rate.csv')
```

```python
df2.head()
```

```python
gdp = df1[' GDP'].tolist()
ur = df2[' Unemployment Rate (%)']
gdp = gdp[::-1]
gdp = gdp[0:21]
ur = ur[::-1]
ur = ur[0:21]
```
